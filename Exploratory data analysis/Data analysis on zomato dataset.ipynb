# removing errors and missing values from rate column

```python
data.rate.unique()
```

# define a function and add conditional loop

```python
import numpy as np
def clean_rate(r):
    if(r=='NEW' or r=='-'):
        return np.nan
    else:
        r=str(r).split('/')
        r=r[0]
        return float(r)
```

# apply the new function to the ratings of data

```python
data.rate=data['rate'].apply(clean_rate)
```

# check if the function is applied

```python
data.rate.unique()
data.info()
```

# replace all the null values in "rate" column with mean of the rate column

```python
data.rate.mean()
data['rate']=data['rate'].fillna(data['rate'].mean())
```

```python
data.cost= data['cost'].str.replace(',','').astype(float)
data.cost.head(10)
```

# Plotting a bar graph for the count of restaurants in a location

```python
#change figure size from default to below given values
plt.figure(figsize=(16,10)) #inches

#plot graph for column "Location"
sns.countplot(x=data['location'])

plt.title('Number of Restaurants in a location')

plt.xticks(rotation=90) 
plt.show()
```

# Get data of top 10 locations with highest number of restaurants

```python
top_locations= data['location'].value_counts().head (10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.index, y=top_locations.values,
palette='viridis')

plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Locations with Maximum Number of Restaurants')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants') 
plt.show()
```

# Count no of restaurants in each rating category

```python
high_rating = (data['rate'] >3.5).sum()
low_rating = (data['rate'] <=3.5).sum()
```

# Plot a bar graph for the rating category

```python
plt.bar(['Higher Rating', 'Lower Rating'], ['high_rating', 'low_rating'],
color= ['teal', 'orange'])

plt.title('Number of restaurants with rating higher or less than 4') 
plt.xlabel("Rating") 
plt.ylabel('Number of restaurants')
plt.show()
```

# plot a bar graph for restaurants with maximum rating

```python
top_rating=data.groupby('name')['rate'].max().sort_values(ascending=False).head(10)
sns.barplot(x = top_rating.index, y = top_rating.values)

plt.xticks(rotation=45, ha ='right')

plt.title('Restaurants with maximum Rating')

plt.xlabel('Restaurants')

plt.ylabel('Number of Rating')

plt.show()
```

# plot a bar graph for 'Name' and 'Votes' column

```python
top_votes = data.groupby('name')['votes'].max().sort_values(ascending=False).head(10)

plt.figure(figsize = (10, 6))

sns.barplot(x=top_votes.index, y=top_votes.values, palette='Paired')

plt.xticks(rotation=45, ha='right')

plt.title('Restaurants with Maximum Votes')

plt.xlabel('Restaurants')

plt.ylabel('Number of Votes')

plt.show()
```

# Same columns but in scatterplot graph

# plotting a histogram using seaborn

```python
top_exp= data.groupby('name')['cost'].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))

sns.histplot(data[data['name'].isin(top_exp.index)], x='cost',bins=20, kde=True, color='pink', element='step', stat='density')
plt.title('Distribution of Costs for Top 10 Expensive Restaurants')
plt.xlabel('Cost')
plt.ylabel('Density')
plt.show()
```
